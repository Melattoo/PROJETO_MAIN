```python
#CONTAR PEDIDOS POR MÃŠS E ORDENAR PARA QUANDO CRIAR O GRAFICO ELE FICAR EM ORDEM CRONOLOGICA

df_olist_orders_dataset['order_purchase_timestamp'] = pd.to_datetime(df_olist_orders_dataset['order_purchase_timestamp'])
df_olist_orders_dataset['mes_ano'] = df_olist_orders_dataset['order_purchase_timestamp'].dt.to_period('M')
vendas_por_mes = df_olist_orders_dataset['mes_ano'].value_counts().sort_index()
print(vendas_por_mes)
```

```python
#CRIAR UM GRFICO PARA DEMONSTRAR A QUANTIDADE
plt.figure(figsize=(12, 6))
vendas_por_mes.plot(kind='line', marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('EvoluÃ§Ã£o Mensal de Pedidos', fontsize=16, pad=20)
plt.xlabel('MÃªs/Ano', fontsize=12)
plt.ylabel('NÃºmero de Pedidos', fontsize=12)
plt.tight_layout() #IMPORTANTE PARA NAO "CORTAR" O GRAFICO
plt.show()
```
